```python
import pandas as pd
import numpy as np
from selenium import webdriver 
import warnings
warnings.filterwarnings("ignore")
import time
import requests

from selenium.common.exceptions import StaleElementReferenceException
from selenium.common.exceptions import ElementNotInteractableException
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import TimeoutException

from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
import re
```

```python

urls=list()
products=['smartphones','laptop','Headphones','Cameras','Printers','Monitors','Home theater','Routers','smart watches']
for product in products:
    
    dr=webdriver.Chrome(r"C:\web driver\New Folder\chromedriverupdated2.exe")
    dr.get('https://www.flipkart.com')
    dr.maximize_window()
    time.sleep(5)

    cancel=dr.find_element("xpath","/html/body/div[2]/div/div/button")
    cancel.click()

    try:
        search=dr.find_element("xpath","/html/body/div[1]/div/div[2]/div[1]/div[2]/div[2]/form/div/div/input")
        search.send_keys(product)
    except NoSuchElementException as e1:
        print(e1)
        search=dr.find_element(by="xpath",value="//input[@class='_3704LK']")
        search.send_keys(product)

    search_button=dr.find_element('xpath',"//button[@class='L0Z3Pu']")
    search_button.click()
    time.sleep(10)


   
    for i in range(5):
        try:
            url_tag=dr.find_element('xpath',"//div[@class='_2kHMtA']/a")

        except NoSuchElementException as e2 :
            print(e2)
            try:
                url_tag=dr.find_element(by="xpath",value="//a[@class='_1fQZEK']")
            except :
                url_tag=dr.find_element("xpath","//div[@class='_4ddWXP']/a")


        urls.append(url_tag.get_attribute('href'))

        next_button=dr.find_elements('xpath',"//a[@class='_1LKTO3']")

        next_button[-1].click()

        time.sleep(5)

    dr.close()
print(len(urls),'\n')
urls 
```
